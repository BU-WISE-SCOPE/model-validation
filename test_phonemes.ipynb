```python
import azure_recognizer as ar
from pathlib import Path
from enum import Enum
```

```python
class PhonemeSet(Enum):
    VOWELS = "AEIOU"
    CONSONANT_CONTINUOUS = "FLMNRSUVZ"
    CONSONTNT_STOP = "BCDGHJKPT"
    W = "W"
    Y = "Y"
    Q = "Q"
    X = "X"

ref_text_map = {
    "A long": "A",
    "E long": "E",
    "I long": "I",
    "O long": "O",
    "U long": "U",
    "A short": "Ah",
    "E short": "Eh",
    "I short": "Eh", # Produces better result than `Ih`
    "O short": "Oh",
    "U short": "Uh",

    "F": "Eff",

    "W": "Woo",
    "Y": "Yee",
    "Q": "Koo",
    "X": "Eks",

}
```

```python
ecri_path = Path("data/ecri_phoneme_data")
ecri_pairs = []
for path in list(ecri_path.rglob("*.wav")):
    transcription = " ".join(path.stem.split(" ")[0:-1])
    print(transcription)
    ecri_pairs.append((path, transcription))
```

## Testing Short Vowel Sounds

```python
for audio_path, transcription in ecri_pairs:
    if "short" in transcription:
        print(f"Processing {transcription}") 
        result = ar.recognize_from_wav(str(audio_path), ref_text_map[transcription])
        ar.display_result(result)
```

## Testing Long Vowel Sounds

```python
for audio_path, transcription in ecri_pairs:
    if "long" in audio_path.stem:
        print(f"Processing {audio_path.stem}")

        result = ar.recognize_from_wav(str(audio_path), transcription)
        ar.display_result(result)
```
